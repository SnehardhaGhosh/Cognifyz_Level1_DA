Task: Descriptive Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load your dataset
df = pd.read_csv('Restaurant.csv')
```

```python

# Select only numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Display the numerical columns
print("Numerical columns in the dataset:\n")
print(num_cols)

# Calculate and display basic statistical measures
print("\nBasic Statistical Measures:\n")
print(df[num_cols].describe())  # gives count, mean, std, min, max, quartiles

# Calculate median separately
print("\nMedian of each numerical column:\n")
print(df[num_cols].median())

# You can also access individual measures like this:
print("\nMean values:\n", df[num_cols].mean())
print("\nStandard deviation values:\n", df[num_cols].std())
print("\nVariance values:\n", df[num_cols].var())
```

```python
# -------------------------------
# 1️⃣ Explore Categorical Variables
# -------------------------------

cat_cols = ['Country Code', 'City', 'Cuisines']  # Add other categorical columns if needed

for col in cat_cols:
    print(f"\n--- Distribution of {col} ---")
    print(df[col].value_counts())  # counts of each category
    print("\nTop 5 most frequent values:")
    print(df[col].value_counts().head(5))  # top 5 values
    
    # Plot countplot
    plt.figure(figsize=(10,4))
    sns.countplot(y=df[col], order=df[col].value_counts().index[:10])
    plt.title(f"Top 10 {col} by Count")
    plt.xlabel("Number of Restaurants")
    plt.ylabel(col)
    plt.show()
```

```python
# -------------------------------
# 2️⃣ Identify Top Cuisines and Cities
# -------------------------------

# Top 10 Cuisines
top_cuisines = df['Cuisines'].value_counts().head(10)
print("\nTop 10 Cuisines with Most Restaurants:")
print(top_cuisines)

# Top 10 Cities
top_cities = df['City'].value_counts().head(10)
print("\nTop 10 Cities with Most Restaurants:")
print(top_cities)

# Optional: Visualize as bar charts
plt.figure(figsize=(10,4))
plt.barh(top_cuisines.index, top_cuisines.values, color=plt.cm.viridis(range(len(top_cuisines))))
plt.title("Top 10 Cuisines by Number of Restaurants")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine")
plt.show()

plt.figure(figsize=(10,4))
plt.barh(top_cuisines.index, top_cuisines.values, color=plt.cm.viridis(range(len(top_cuisines))))
plt.title("Top 10 Cities by Number of Restaurants")
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.show()
```
