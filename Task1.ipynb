Explore the dataset and identify the number
of rows and columns

```python
# Import the required library
import pandas as pd

# Load the dataset (replace 'your_dataset.csv' with your actual file name)
df = pd.read_csv('Restaurant.csv')

# Display the first few rows (optional)
df.head()
```

Explore the dataset and identify the number
of rows and columns.


```python

# Display the first few rows (optional, just to explore)
print("First 5 rows of the dataset:")
print(df.head())

# Identify the number of rows and columns
rows, cols = df.shape
print("\nNumber of rows:", rows)
print("Number of columns:", cols)
```

Check for missing values in each column and
handle them accordingly.

```python
# Check for missing values
print("Missing values before handling:")
print(df.isnull().sum())

# Handle missing values safely (no inplace=True)
for column in df.columns:
    if df[column].dtype == 'object':  # categorical
        df[column] = df[column].fillna(df[column].mode()[0])
    else:  # numerical
        df[column] = df[column].fillna(df[column].mean())

# Verify again
print("\nMissing values after handling:")
print(df.isnull().sum())
```

Data Type Conversion

```python
# Check current data types
print("Data types before conversion:\n")
print(df.dtypes)

# Example: Convert columns if needed
# (You can modify these lines based on your dataset)
# Suppose 'Aggregate rating' should be numeric
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')

# Example: If there are date columns
# df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Verify again
print("\nData types after conversion:\n")
print(df.dtypes)
```

Analyze the Distribution of the Target Variable (“Aggregate rating”)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Display basic statistics of Aggregate rating
print("Summary of Aggregate rating:\n")
print(df['Aggregate rating'].describe())

# Plot the distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Aggregate rating'], bins=10, kde=True)
plt.title("Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.show()

# Check value counts for each rating (to detect class imbalance)
print("\nValue counts of Aggregate rating:")
print(df['Aggregate rating'].value_counts())

# Plot value counts as bar chart
plt.figure(figsize=(8, 5))
sns.countplot(x='Aggregate rating', data=df)
plt.title("Class Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Records")
plt.show()
```
